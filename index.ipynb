Create dummy variables for the six categorical features you chose remembering to drop the first. Drop the categorical columns that you used, concat the dummy columns to our continuous variables and asign it to a new variable `ames_preprocessed`

```python
# Create dummy variables for your six categorical features
category = df.SaleType.unique()
category_series = pd.Series(category)
cat_origin = category_series.astype('category')
cat_origin
```

```python
cat_origin.cat.codes
```

```python
from sklearn.preprocessing import LabelEncoder
lb_make = LabelEncoder()

origin_encoded = lb_make.fit_transform(cat_origin)
origin_encoded
```

```python
from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()
origin_dummies = lb.fit_transform(cat_origin)
# You need to convert this back to a dataframe
origin_dum_df = pd.DataFrame(origin_dummies,columns=lb.classes_)
origin_dum_df
```

```python
df_cat_encoded = df.drop('SaleType', axis=1)
df_cat_encoded = pd.concat([df_cat_encoded, origin_dum_df], axis=1)
df_cat_encoded.head()
```

```python
categorial = ['BldgType', 'KitchenQual', 'SaleType', 'MSZoning', 'Street', 'Neighborhood']
dummies = pd.get_dummies(df[categorial], prefix=categorial, drop_first=True)
dummies
```

```python
df_preprocessed = df.drop(categorial, axis=1)
df_preprocessed = pd.concat([df_preprocessed, dummies], axis=1)
df_preprocessed.head()
```
